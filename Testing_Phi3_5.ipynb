```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, BitsAndBytesConfig
```

```python
import torch
import getpass
import os

device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
device
```

```python
torch.random.manual_seed(42)
```

```python
os.environ["HF_TOKEN"] = getpass.getpass()
```

```python
id_model="microsoft/Phi-3.5-mini-instruct"
```

```python
model = AutoModelForCausalLM.from_pretrained(
    id_model,
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
    attn_implementation="eager"
    )
```

```python
tokenizer = AutoTokenizer.from_pretrained(id_model)
```

```python
pipe = pipeline("text-generation",model = model ,tokenizer=tokenizer)
```

```python
generation_args= {

    "max_new_tokens" : 500,
    "return_full_text" :False,
    "temperature" : 0.2,
    "do_sample": True,

}
```

For my goal of leaning more about Siberian Huskyes, the "temperature" parameter on 0.3 gave me a better result.

```python
prompt = "Explique as difereças de linhagem dos Huskys Siberianos"
output = pipe(prompt,**generation_args)
```

```python
print(output[0]['generated_text'])
```
